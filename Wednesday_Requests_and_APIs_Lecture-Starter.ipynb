```python
class CurrentTemp:
    def __init__(self, city_name, state_name, temp, condition, feels_like):
        self.city = city_name
        self.state = state_name
        self.temp = temp
        self.condition = condition
        self.feels_like = feels_like
        
    def __str__(self):
        degree_sign = u'\N{DEGREE SIGN}'
        return f"{self.city}, {self.state} - {self.temp}{degree_sign}F"
    
    def __repr___(self):
        return f'<Weather {self.city}| {self.temp}{degree_sign}F'
    
city_weather = CurrentTemp(city, state, current_temp, current_condition, current_feels_like)

print(city_weather)
```

```python
print(city_weather.feels_like)
print(city_weather.condition)
```

```python
class WeatherAPI:
    
    def __init__(self, api_key):
        self.api_key = api_key
        self.base_url = 'http://api.weatherapi.com/v1'
        
    def _get(self, q, api_method, **kwargs):
        request_url = f"{self.base_url}{api_method}?key={self.api_key}&q={q}"
        days = kwargs.get('days')
        if days:
            request_url += f"&days={days}"
        weather_response = requests.get(request_url)
        if weather_response.ok:
            return weather_response.json()
        else:
            print('There was an error')
    
    def get_current_weather(self, city):
        weather_data = self._get(city, '/current.json')
        city = weather_data['location']['name']  
        state = weather_data['location']['region']
        current_temp = weather_data['current']['temp_f']
        current_condition = weather_data['current']['condition']['text']
        current_feels_like = weather_data['current']['feelslike_f']
        city_temp = CurrentTemp(city, state, current_temp, current_condition, current_feels_like)
        return city_temp
    
    def get_forecast_weather(self, city, days=3):
        weather_data = self._get(city, '/forecast.json', days = days)
        city = weather_data['location']['name']  
        state = weather_data['location']['region']
        list_of_days = weather_data['forecast']['forecastday']
        for day in list_of_days:
            date = day['date']
            high = day['day']['maxtemp_f']
            low = day['day']['mintemp_f']
            condition = day['day']['condition']['text']
            print(f"On {date}, it will be {condition} with a high of {high} and a low of {low}")
            
            
        return weather_data
        
        
```

```python
def main():
    api_key = input('What is your api key? ')
    client = WeatherAPI(api_key)
    while True:
        city = input('For which city woulld you like to check the weather? or type \'quit\' to exit')
        if city.lower() == 'quit':
            break
        method = input('Are you looking for the current or forecast? ')
        while method not in {'current', 'forecast'}:
            method = input('Please enter either current or forecast')
        if method == 'current':
            city_weather = client.get_current_weather(city)
            print(f"It is currently {city_weather.temp} and {city_weather.condition}. \
If feels like it is {city_weather.feels_like}")
        else:
            days = input('For how many days in the future are you interested (limit 3) ')
            client.get_forecast_weather(city, days)
            
            
main()
```
